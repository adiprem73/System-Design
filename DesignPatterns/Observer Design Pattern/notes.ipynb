# Observer Design Pattern


--> the example is of youtube notification system. when we subscribe a channel, we get notification about any new video upload. hence this is observer design pattern

## Polling Technique
the object which needs to be observed (youtube channel) is called subject

jiski state janni hai wo subject or observable. and the object which wnats to know the state is the observer. 

the observer sends repitive request to the observable in some frequency. which is very bad for system

### hence we go for Pushing

## Pushing

in this the observable sends informtion to the observers that its state has beeen changed

- so the observable or channel must keep a list of all observers
