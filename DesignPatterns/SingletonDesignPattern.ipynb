# SingleTon Design Pattern

SingleTon is a class which allows only one object opr isntance to be created. if we try to makke naother object, it returns that class only.

there are two types of memories used in the backend.
1. Stack memory which stores all the primitive data types like int, char etc
2. Heap memory which stores all the classess etc

- creates a private constructor
- creates a static instance (getInstance()) that returns the same instance every time

Practical Use Cases:
1. Logging Systems
2. Database Connection
3. Configuration Manager

We should no use singleton in scenarios where we know that we need multiple instances.
