## Strategy Design Pattern
it defnies a family of algorithm that we put into sperate classess so that they can be changed at run time

the whole idea being design pattern is to seperate the coide into two sectons.
1. code which needs to be changed  as the code is developed more and more
2. code which can stay the same 

for example, in the robot class, projection() is the code which does nto need to chnage even in the subclasses. this is becase each subclass has its own looks so that will all have projections. 

the issue comes with the other functiosn talk(), walk(), fly(),,, as some classes will talk walk or fly and some would not. hence we take them out of the robot classe
