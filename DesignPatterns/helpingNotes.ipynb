Strategy design patterns:

- Strategy Interface
    - Common interface for all algorithms.
- Concrete Strategies
    - Different implementations of the algorithm.
- Context
    - Uses a strategy object to execute behavior.
- Client
    - Chooses which strategy to use.

पहले तीन strategy interfaces बनाएँगे (Talkable, Walkable, Flyable) जिनमें virtual functions होंगे। फिर हम इन interfaces को inherit करके अलग-अलग strategies बनाएँगे—जैसे NormalWalk या NoWalk। आखिर में, Robot class बनेगी, जिसमें ये तीनों strategies को inject करेंगे। अब runtime पर, तुम कोई भी concrete strategy चुन सकते हो—robot उसी हिसाब से react करेगा!
